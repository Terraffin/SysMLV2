```sysml
pkg SyringeSystem {

 

    // Items
    item def Drug {
        value type: String[1]; // type or name of the drug
        value volume: Real[1]; // volume of the drug
    }

 

    // Parts
    part def Barrel {
        value capacity: Real[1]; // volume capacity of the barrel
        feature drugPort: Feature[1]<Drug>; // feature for drug flow
    }

    part def Plunger {
        value position: Real[1]; // position of the plunger in the barrel
    }

    part def Needle {
        attribute length: Real[1]; // length of the needle
        attribute diameter: Real[1]; // diameter of the needle
        item drugInput: Feature[1]<Drug>; // feature for drug input
        item drugOutput: Feature[1]<Drug>; // feature for drug output
    }

 

    part def DrugVial {
        value volumeAvailable: Real[1]; // volume of drug available in the vial
        feature drugOutput: Feature[1]<Drug>; // feature for drug output
    }

    part def Syringe {
        part barrel: Barrel;
        part plunger: Plunger;
        part needle: Needle;
        needle.drugInput => barrel.drugPort; // connection from needle to barrel
    }

 

    // Stream Connection
    stream def DrugFlow {
        item drug: Drug;
        from src: Feature[1]<Drug>;
        to dst: Feature[1]<Drug>;
    }

 

    // Actions and Interactions
    action def drawDrug {
        entry;
        do {
            // Create a stream connection from vial to syringe
            stream DrugFlow from DrugVial.drugOutput to Syringe.needle.drugInput;
        }
        exit;
    }

 

    action def inject {
        entry;
        do {
            // Create a stream connection from syringe to outside
            stream DrugFlow from Syringe.needle.drugOutput;
        }
        exit;
    }

 

    seq def operateSyringe {
        msg drawDrugFromVial(drugVial, needle);
        msg expelFluid(barrel, plunger, needle);
    }

 

    // States
    state def SyringeState {
        state empty;
        state drawingDrug;
        state filled;
        state expelling;

        trans initial to empty;
        trans empty to drawingDrug on drawDrugFromVial;
        trans drawingDrug to filled on exit;
        trans filled to expelling on expelFluid;
        trans expelling to empty on exit;
    }

 

} // end of pkg SyringeSystem
```
